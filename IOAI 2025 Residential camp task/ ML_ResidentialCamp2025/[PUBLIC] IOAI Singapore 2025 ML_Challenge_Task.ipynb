```python
from datetime import datetime
import pytz
import pandas as pd
import plotly.express as px


from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier, CatBoostRegressor

from sklearn.metrics import accuracy_score, confusion_matrix

from sklearn.model_selection import train_test_split



# Define the timezone
timezone = pytz.timezone('Asia/Singapore')  # GMT+8


### DO NOT MODIFY OR EDIT THIS FUNCTION
def train_test(name, model, X_train, X_val, y_train, y_val, X_test=None):

  print("=" * 50)
  print("Training and Testing", name)
  print("=" * 50)
  # Model Training
  model.fit(X_train, y_train)

  # Make predictions
  y_pred = model.predict(X_val)

  # Evaluate the model
  # Calculate evaluation metrics
  accuracy = accuracy_score(y_val, y_pred)
  print(f'Accuracy: {accuracy}')

  ## Display Confusion Matrix
  show_confusion_matrix(y_val, y_pred, name)

  # Get the current time in UTC+8
  utc_now = datetime.now(pytz.utc)
  current = utc_now.astimezone(timezone)
  current_datetime = current.strftime("%Y-%m-%d_%H-%M-%S")
  print(f'Time: {current_datetime}')

  # OUTPUT predicted label into csv file
  if X_test is not None:
    y_test = model.predict(X_test)
    pd.DataFrame(y_test).to_csv(f"{youremail}_{current_datetime}_{name}_{current.timestamp()}_.csv")

  print("=" * 50)
  ## VALIDATION RESULT
  return {"name": name, "Acc": accuracy, "timestamp": current.timestamp()}, y_pred


def show_confusion_matrix(y_test, y_pred, name=""):

  class_names = sorted(y_test.unique())  # Get unique class names (e.g., ['Class1', 'Class2'])

  mat = confusion_matrix(y_test, y_pred)
  px.imshow(mat,
            x=class_names,
            y=class_names,
            title=f"{name} : Confusion matrix between true and predicted label.",
            text_auto=True,
            color_continuous_scale="turbo",
            labels=dict(x='y_test', y='y_pred', color='Value'),
            width=1000,
            height=1000,
            ).show()

def predict_and_generate_output_file(model, X_test):
  utc_now = datetime.now(pytz.utc)
  current = utc_now.astimezone(timezone)
  current_datetime = current.strftime("%Y-%m-%d_%H-%M-%S")
  print(f'Time: {current_datetime}')

  ## Predict
  y_test = model.predict(X_test)

  ## Saving Output File.
  pd.DataFrame(y_test).to_csv(f"MLTask_{youremail}_{yourname}_{current_datetime}_.csv")
```
